Load Libraries 

```python
import pandas as pd
from pyvis.network import Network
import networkx as nx
```

```python
gbmSYGNAL=pd.read_csv('zip.data/data/gbmSYGNAL.csv', header=0, index_col=0).dropna()
print(gbmSYGNAL.shape)
```

```python
gbmSYGNAL['Genes'] = [i.split('|') for i in gbmSYGNAL['Genes']]
gbmSYGNAL
```

```python
gbmSYGNAL['miRNAs'] = [i.split('|') for i in gbmSYGNAL['miRNAs']]
gbmSYGNAL
```

```python
gbmSYGNAL['Enriched GO BP Terms'] = [i.split('|') for i in gbmSYGNAL['Enriched GO BP Terms']]
gbmSYGNAL
```

```python
hallmarks = ['Sustained Angiogenesis', 'Insensitivity To Antigrowth Signals', 'Evading Apoptosis', 'Limitless Replicative Potential', 'Evading Immune Detection', 'Tissue Invasion And Metastasis', 'Self Sufficiency In Growth Signals', 'Tumor Promoting Inflammation', 'Reprogramming Energy Metabolism', 'Genome Instability And Mutation']
def getHallmarks(scores, cutoff=0.8):
    return list((scores[scores>cutoff]).index)
```

identify which hallmark is significant for which clusters 

```python
clusterHallmarks = []
for cluster in gbmSYGNAL.index:
    scores = gbmSYGNAL.loc[cluster, hallmarks]
    clusterHallmarks.append(getHallmarks(scores))

clusterHallmarks
```

DataFrame as a new column 'sigHallmarks':

```python
gbmSYGNAL['sigHallmarks'] = clusterHallmarks
gbmSYGNAL
```

Now we can subset the network
down to those with an association with a hallmark of cancer:

```python
gbmSYGNAL_hm = gbmSYGNAL.loc[[True if len(gbmSYGNAL.loc[cluster,'sigHallmarks'])>0 else False for cluster in gbmSYGNAL.index]]
gbmSYGNAL_hm
```

 miRNA  cluster  GO term  hallmark of cancer

```python
miRNA_HM_dict={}
for cluster in gbmSYGNAL.index: 
    for miRNA in gbmSYGNAL.loc[cluster, 'miRNAs']: 
        if not miRNA in miRNA_HM_dict: 
            miRNA_HM_dict[miRNA]=[]
        for hm1 in gbmSYGNAL.loc[cluster, 'sigHallmarks']:
            if not hm1 in miRNA_HM_dict: 
                miRNA_HM_dict[miRNA].append(hm1)

print(miRNA_HM_dict)
len(set(miRNA_HM_dict.keys()))

meaningfulmiRNA=[]
for miRNA in miRNA_HM_dict:
    if len(miRNA_HM_dict[miRNA])>1:       
        meaningfulmiRNA.append(miRNA)

print(meaningfulmiRNA)

print(len(meaningfulmiRNA))
```

```python
miRNAs_biclus = {}
for cluster in gbmSYGNAL.index:
    for miRNA in gbmSYGNAL.loc[cluster,'miRNAs']:
        for miRNA_impo in meaningfulmiRNA:
            if not miRNA_impo in miRNAs_biclus:
                miRNAs_biclus[miRNA_impo] = []
            miRNAs_biclus[miRNA_impo].append(cluster)
print(miRNAs_biclus)
#print(len(miRNAs_biclus.keys()))

meaningfulBiclus=[j for i in miRNAs_biclus for j in miRNAs_biclus[i]]
print(len(set(meaningfulBiclus)))
```

```python
lnk_miRNAs = {}
lnk_go_miRNAs = {}
hm_dict = {}
go_dict = {}

for miRNA in miRNAs_biclus:
# Idenitfy which biclusters have same go term -> hallmark relationships
    # First, fill hm_dict with hallmarks and go_dict with go terms
    for cluster in miRNAs_biclus[miRNA]:
        # Build clusters so can test for common hallmarks
        for hm1 in gbmSYGNAL.loc[cluster,'sigHallmarks']:
            if not hm1 in hm_dict:
                hm_dict[hm1] = []
            hm_dict[hm1].append(cluster)
        # Build clusters so can test for common GO terms
        for go1 in gbmSYGNAL.loc[cluster,'Enriched GO BP Terms']:
            if not go1 in go_dict:
                go_dict[go1] = []
            go_dict[go1].append(cluster)
        # Second, find hallmarks in common across clusters, and refine subnetwork (cc_miRNAs)
        for hm1 in hm_dict:
            if not miRNA in lnk_miRNAs:
                lnk_miRNAs[miRNA] = {}
            lnk_miRNAs[miRNA][hm1] = miRNAs_biclus[miRNA]
        # Thrid, find go terms in common across clusters, and refine subnetwork
        for go1 in go_dict:
            if miRNA in lnk_miRNAs:
                if not miRNA in lnk_go_miRNAs:
                    lnk_go_miRNAs[miRNA] = {'hallmarks':lnk_miRNAs[miRNA].keys(), 'clusters':miRNAs_biclus[miRNA], 'GO_terms':[]}
                lnk_go_miRNAs[miRNA]['GO_terms'].append(go1)
print(lnk_miRNAs)
print(lnk_go_miRNAs)
```

Build the network #1:  miRNA  bicluster  GO term  hallmark of cancer
First need to initialize a networkx directed graph:

```python
gbmCancer = nx.DiGraph()
```

```python
for miRNA in lnk_go_miRNAs:
    for cluster in lnk_go_miRNAs[miRNA]['clusters']:
        # Add miRNA -> cluster
        gbmCancer.add_edge(miRNA,cluster)
        gbmCancer.nodes[cluster]['group'] = 'bicluster'
        gbmCancer.nodes[miRNA]['group'] = 'miRNA'
        # Add cluster -> GO term -> HM
        for go1 in lnk_go_miRNAs[miRNA]['GO_terms']:
            gbmCancer.add_edge(cluster,go1)
            gbmCancer.nodes[go1]['group'] = 'go_term'
            for hm1 in lnk_go_miRNAs[miRNA]['hallmarks']:
                gbmCancer.add_edge(go1,hm1)
                gbmCancer.nodes[hm1]['group'] = 'hallmark'
```

```python
# Number of nodes: 

print(len(meaningfulmiRNA)) #miRNA
print(len(set(meaningfulBiclus))) #biclusters 
print(len(set(go_dict)))  #GO terms
print(len(set(hm_dict)))  #hallamrk
```

```python
gbm_nt = Network('800px','800px', directed=True, notebook=True)
#gbm_nt.show_buttons(filter_=['physics'])
#gbm_nt.hrepulsion()
gbm_nt.from_nx(gbmCancer)
gbm_nt.show('gbmCancer_wGOTermsClusHM_EL.html')
```

Build the network #2: miRNA  cluster  hallmark of cancer

```python
gbmCancer = nx.DiGraph()
```

```python
for miRNA in lnk_go_miRNAs:
    for cluster in lnk_go_miRNAs[miRNA]['clusters']:
        # Add miRNA -> cluster
        gbmCancer.add_edge(miRNA,cluster)
        gbmCancer.nodes[cluster]['group'] = 'bicluster'
        gbmCancer.nodes[miRNA]['group'] = 'miRNA'
        # Add cluster -> HM
        for hm1 in lnk_go_miRNAs[miRNA]['hallmarks']:
            gbmCancer.add_edge(cluster,hm1)
            gbmCancer.nodes[hm1]['group'] = 'hallmark'
```
